# Basic Plots using Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Generation of X axis and random number for y axis

```python
x = np.linspace(1,100,100,dtype='int8')#linspace retunrs evenly spaced samples
y1 = np.random.randint(1,100,100)
y2 = np.random.randint(1,100,100)
y3 = np.random.randint(1,100,100)
```

# Line Plot

```python
plt.plot(x,y1)
```

# Modify figure size

```python
plt.figure(figsize=(10,5))
plt.plot(x,y1)
```

# Scatter Plot

```python
plt.scatter(x,y1,alpha=.9,color="red") # alpha is for transparency
```

# Bubble Plot

```python
plt.figure(figsize=(16,4))
for idx in range(len(x)):
    plt.scatter(x[idx],y1[idx],alpha=.5,color="blue",s=.05*np.pi*y1[idx]**2) # s = size
```

# Vertical Bar Plot

```python
plt.figure(figsize=(16,4))
plt.bar(x[:10],y1[:10],width=.7,color="green")
```

# Horizontal Bar Plot

```python
plt.figure(figsize=(30,10))
plt.barh(x[:10],y1[:10],color="orange")
```

# Stacked Bar plot

```python
plt.figure(figsize=(16,4))
plt.bar(x[:20],y1[:20],width=.8)
plt.bar(x[:20],y2[:20],width=.8,bottom=y1[:20])
plt.bar(x[:20],y3[:20],width=.8,bottom=y1[:20]+y2[:20])
```

# Histogram

```python
plt.figure(figsize=(16,4))
plt.hist(y1,rwidth=.9,bins=10,color="indigo",align="mid")
```
